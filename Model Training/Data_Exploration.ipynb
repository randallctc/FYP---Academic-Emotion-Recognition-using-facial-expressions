```python
import pandas as pd
import os
```

```python
df = pd.read_csv("AllLabels_binary.csv")
```

```python
emotion_cols = ["Boredom", "Engagement", "Confusion", "Frustration"]

total_clips = len(df)
print(f"\nTotal Clip: {total_clips} ")

print("\nEmotion Counts (Individual):")
for col in emotion_cols:
    count = df[col].sum()
    print(f"{col}: {count} clips")

df["NumEmotions"] = df[emotion_cols].sum(axis=1)

no_emotion_count = (df["NumEmotions"] == 0).sum()
single_emotion_count = (df["NumEmotions"] == 1).sum()
mixed_emotion_count = (df["NumEmotions"] >= 2).sum()

print("\nEmotion Combination Summary:")
print(f"Clips with NO emotions:       {no_emotion_count}")
print(f"Clips with ONE emotion:       {single_emotion_count}")
print(f"Clips with MIXED emotions:    {mixed_emotion_count}")

print("\nMixed Emotion Breakdown (# of emotions per clip):")
print(df["NumEmotions"].value_counts().sort_index())

print("\nHow often each emotion appears in MIXED clips:")
mixed_df = df[df["NumEmotions"] >= 2] 
for col in emotion_cols:
    count = mixed_df[col].sum()
    print(f"{col}: {count} clips")
```

```python
df["ClipID_noext"] = df["ClipID"].astype(str).str.strip().str.replace(r"\.avi$|\.mp4$|\.mov$", "", regex=True)
df["NumEmotions"] = df[emotion_cols].sum(axis=1)
BASE_DIR = r"D:\FYP\DAiSEE\Data"
splits = ["Train", "Test", "Validation"]

def summarize_split(split_name):
    path = os.path.join(BASE_DIR, split_name)
    video_ids = []
    for root, dirs, files in os.walk(path):
        for f in files:
            if f.lower().endswith((".avi", ".mp4", ".mov")):  # handle all video extensions
                video_ids.append(os.path.splitext(f)[0])  # remove extension
    split_df = df[df["ClipID_noext"].isin(video_ids)]
    
    print(f"===== {split_name.upper()} SUMMARY =====")
    total = len(split_df)
    print(f"Total clips: {total}")
    
    print("\nEmotion Counts (Individual):")
    for col in emotion_cols:
        print(f"{col}: {split_df[col].sum()} clips")

    print("\nEmotion Combination Summary:")
    print(f"No emotion (0):    {(split_df['NumEmotions']==0).sum()}")
    print(f"One emotion (1):   {(split_df['NumEmotions']==1).sum()}")
    print(f"Mixed emotions:    {(split_df['NumEmotions']>=2).sum()}")

    print("\nMixed Emotion Breakdown (# of emotions per clip):")
    print(split_df["NumEmotions"].value_counts().sort_index())

    mixed = split_df[split_df["NumEmotions"] >= 2]

    print("\nHow often each emotion appears in MIXED clips:")
    for col in emotion_cols:
        print(f"{col} appears in mixed clips: {mixed[col].sum()}")

    return split_df

# Run summary for each split
train_df = summarize_split("Train")
test_df = summarize_split("Test")
val_df = summarize_split("Validation")

print("\n\nDONE.")
```

```python
df = pd.read_csv("AllLabels_binary.csv")
```

```python
df["ClipID_noext"] = (
    df["ClipID"]
    .astype(str)
    .str.strip()
    .str.replace(r"\.avi$|\.mp4$|\.mov$", "", regex=True)
)

df["NumEmotions"] = df[emotion_cols].sum(axis=1)

BASE_DIR = r"D:\FYP\DAiSEE\Data"
splits = ["Train", "Test", "Validation"]

def summarize_split_single_emotion(split_name):
    path = os.path.join(BASE_DIR, split_name)
    video_ids = []

    for root, dirs, files in os.walk(path):
        for f in files:
            if f.lower().endswith((".avi", ".mp4", ".mov")):
                video_ids.append(os.path.splitext(f)[0])

    # Match clips in this split
    split_df = df[df["ClipID_noext"].isin(video_ids)]

    # 🔹 KEEP ONLY clips with exactly ONE emotion
    single_df = split_df[split_df["NumEmotions"] == 1]

    print(f"===== {split_name.upper()} (SINGLE-EMOTION ONLY, BINARY) =====")
    total = len(single_df)
    print(f"Total clips (1 emotion only): {total}")

    print("\nEmotion Counts:")
    for col in emotion_cols:
        print(f"{col}: {single_df[col].sum()} clips")

    print("\nEmotion Distribution Check:")
    print(single_df["NumEmotions"].value_counts())

    return single_df


# Run summary for each split
train_single_df = summarize_split_single_emotion("Train")
test_single_df  = summarize_split_single_emotion("Test")
val_single_df   = summarize_split_single_emotion("Validation")

print("\n\nDONE.")
```
